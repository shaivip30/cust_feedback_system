```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
from transformers import pipeline
from datetime import datetime

# Pretrained Transformer Model
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Store chat history
history = []

def analyze_text(text):

    result = sentiment_pipeline(text)[0]

    sentiment = result["label"]
    confidence = round(result["score"] * 100, 2)

    if sentiment == "POSITIVE":
        emoji = "😊"
        response = "Your text expresses a positive sentiment."
    else:
        emoji = "😔"
        response = "Your text expresses a negative sentiment."

    history.append({
        "Time": datetime.now().strftime("%H:%M:%S"),
        "Text": text,
        "Sentiment": sentiment,
        "Confidence": confidence
    })

    return f"""
{emoji} Sentiment: {sentiment}

🎯 Confidence: {confidence}%

💬 Analysis:
{response}
"""

def generate_chart():

    if len(history) == 0:
        return None

    df = pd.DataFrame(history)

    counts = df["Sentiment"].value_counts()

    plt.figure(figsize=(5,4))
    counts.plot(kind="bar", color=["green","red"])
    plt.title("Sentiment Distribution")
    plt.ylabel("Count")

    return plt.gcf()

with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown(
    """
    # 🤖 AI-Powered NLP Sentiment Analysis System

    ### Features
    - Transformer-based NLP
    - Real-time Sentiment Analysis
    - Confidence Scoring
    - Historical Analytics
    - Interactive Dashboard
    """
    )

    with gr.Row():

        with gr.Column():
            text_input = gr.Textbox(
                label="Enter Text",
                lines=4,
                placeholder="Type your review, feedback, or message..."
            )

            analyze_btn = gr.Button("Analyze", variant="primary")

            output = gr.Textbox(
                label="Analysis Result"
            )

        with gr.Column():
            chart_output = gr.Plot(
                label="Analytics Dashboard"
            )

    analyze_btn.click(
        fn=analyze_text,
        inputs=text_input,
        outputs=output
    ).then(
        fn=generate_chart,
        outputs=chart_output
    )

demo.launch()
```
